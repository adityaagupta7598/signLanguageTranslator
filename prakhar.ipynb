```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier('sign_language-A-Classifier.xml')
```

```python
while True:
    ret,photo = cap.read()
#    grey_image = cv2.cvtColor(photo,cv2.COLOR_BGR2HSV)
#    hands= hand_cascade.detectMultiScale(grey_image)
#    for(x,y,w,h) in hands:
#        cv2.rectangle(grey_image, (x,y), (x+w,y+h), (255,0,0), 5)
    myphoto=photo[5:250, 5:250]
    grey_image=cv2.cvtColor(photo,cv2.COLOR_BGR2GRAY)
    cv2.rectangle(photo,(5,5),(250,250),(255,0,0),5)
    hsv_image=cv2.cvtColor(photo,cv2.COLOR_BGR2HSV)
    hsv_image=cv2.cvtColor(myphoto,cv2.COLOR_BGR2HSV)
    lower_green=np.array([0,20,50])
    upper_green=np.array([20,250,250])
    mask=cv2.inRange(hsv_image, lower_green, upper_green)
    gesture = face_cascade.detectMultiScale(myphoto)
    print(gesture)
    if gesture is ():
        pass
        cv2.imshow('hi' , photo)
        if cv2.waitKey(1)==27:
            break
    else:
        for(x,y,w,h) in gesture:
            cv2.rectangle( photo , (x,y) , (x+w,y+h) , (255,0,0) , 5 )
            cv2.putText(photo, "It's A", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
            cv2.imshow('hi' , photo)
            cv2.imshow('hey',mask)
            if cv2.waitKey(1) == 27:
                break

#    cv2.rectangle(grey_image,(x,y),(x+w,y+h),(255,0,0),5)
#    cv2.rectangle(hsv_image,(x,y),(x+w,y+h),(255,0,0),5)
#    cv2.imshow('hi',photo)
#    cv2.imshow('hey',mask)
#    if cv2.waitKey(1)==27:
#        break
cv2.destroyAllWindows()
cap.release()
```
